```python
with open(r'C:\Users\George\Desktop\List of postal codes of Canada_ M - Wikipedia.html',encoding='utf-8') as html_file:
    wiki= BeautifulSoup(html_file,'lxml')
```

```python
main_body = wiki.body.find(id='content',role='main')
```

```python
main_body = main_body.find(id='bodyContent')
```

```python
main_body = main_body.find(id='mw-content-text')
```

```python
main_body = main_body.find(class_='mw-parser-output')
```

```python
main_body = main_body.find(name='table')
```

```python
main_body = main_body.tbody
```

```python
main_body
```

```python
tr = list(filter(lambda x:x!='\n',main_body.contents))
```

```python
# define the dataframe columns
column_names = ['PostalCode','Borough', 'Neighborhood'] 

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)
```

```python
neighborhoods
```

```python
for k in range(len(tr)):
    td = list(filter(lambda x:x!='\n',tr[k].contents))
    for j in range(len(td)):
        if td[j].i==None:
            PostalCode = td[j].p.b.string
            area = td[j].find_all(name='a')
            Borough = '';
            Neighborhood = ''
            for j in range(len(area)):
                if j == 0:
                    Borough = area[j].string
                if j == 1:
                    Neighborhood = area[j].string
                if j > 1:
                    Neighborhood = Neighborhood +','+area[j].string
        #print(PostalCode)
        #print(Borough)
        #print(Neighborhood)
            neighborhoods = neighborhoods.append({'PostalCode':PostalCode,'Borough':Borough,'Neighborhood':Neighborhood},ignore_index=True)
```
